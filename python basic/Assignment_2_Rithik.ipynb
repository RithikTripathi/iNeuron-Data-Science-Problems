##### Question 10. How can you tell the difference between break and continue?

Solution: Break & Continue statements are basically used to alter the normal flow of a loop.
    Break: Break is used to stop the flow of statement when a particular condition is met.
    Continue: Continue is used to skip the execution of step if a particular condition is met.


    for i in range(0,10) :
        if ( i == 5 ) : 
            break
        else : 
            print(i)
            
output : 1, 2, 3, 4, 5
execution of loop will stop when i = 5

    for i in range(0,10) :
        if ( i == 5 ) : 
            continue
        else : 
            print(i)
            
output : 1, 2, 3, 4, 6, 7, 8, 9
loop will skip printing 5 and will print all other values

##### Question 11. In a for loop, what is the difference between range(10), range(0, 10), and                        range(0, 10, 1)?
